```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


url = 'http://www.aero.iitb.ac.in/~prabhu/tmp/sslc1.csv'
df = pd.read_csv(url, sep=";",na_values=["AA"])
df.head()
```

```python
#Ques 1
list=[]
reg=["A","B","C","D","E","F"]
df_num=df.groupby('region')
for i in reg:
    list.append(len(df_num.get_group(i)))
plt.pie(list,labels=reg,autopct='%.2f%%')
```

```python
#Ques2
list2=[]
condition2=df['pass']=="P"
df_pass=df[condition2].groupby('region')
for i in reg:
    list2.append(float(len(df_pass.get_group(i)))/float(len(df_num.get_group(i))))
    
plt.pie(list2,labels=reg,autopct='%.2f%%')
```

```python
#Ques3

list3=[]
def isNaN(num):
    return ~(num != num)
condition3= isNaN(df['sl']) & isNaN(df['fl']) & isNaN(df['math']) & isNaN(df['sci']) & isNaN(df['ss'])
df_present=df[condition3].groupby('region')
for i in reg:
   list3.append(float(len(df_present.get_group(i)))/float(len(df_num.get_group(i))))

plt.pie(list3,labels=reg,autopct='%.4f%%')
```

```python
#Ques4
df.hist(column = 'total', by='region',figsize=[15,20])
```

```python
#Ques5
df_math = pd.concat([df['math'],df['region']],axis=1,keys=["math","region"])
df_math.boxplot(by="region")
plt.ylabel("Marks")
```

```python
#Ques6
list6=[]
condition6=df['sci']>90
df_sci=df[condition6].groupby('region')
for i in reg:
    list6.append(float(len(df_sci.get_group(i)))/float(len(df_num.get_group(i)))*100)
    
plt.pie(list6,labels=reg,autopct='%.2f%%')
```
